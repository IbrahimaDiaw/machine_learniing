```python
import pandas as pd
import numpy as np
import scipy as py
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline
matplotlib.rcParams['savefig.dpi'] = 144
plt.rcParams['figure.figsize'] = 14, 10

import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
# lecture des fichiers csv
# source des donnees https://wwwbis.sidc.be/silso/datafiles
tache_solaire_jour = pd.read_csv('sunspots/SIDC-SUNSPOTS_D.csv')
tache_solaire_mois = pd.read_csv('sunspots/SIDC-SUNSPOTS_13.csv')
```

```python
# Visualisation de donnees tache solaires importees par jour
tache_solaire_jour.Date = pd.to_datetime(tache_solaire_jour.Date)
tache_solaire_jour.index = tache_solaire_jour.Date
tache_solaire_jour.index.name = None
tache_solaire_jour.sort_values('Date', ascending=True, inplace = True)
tache_solaire_jour.head()
```

```python
# Visualisation de donnees importees 
tache_solaire_mois.Date = pd.to_datetime(tache_solaire_mois.Date)
tache_solaire_mois.index = tache_solaire_mois.Date
tache_solaire_mois.index.name = None
tache_solaire_mois.sort_values('Date', ascending=True, inplace = True)
tache_solaire_mois.head()
```

```python
# le diagramme des taches solaires par jour
tache_solaire_jour['Daily Sunspot Number'].plot(figsize=(12,8), title= 'Activit quotidienne des taches solaires', fontsize=12)
plt.xlabel("Series Temporelles (1818-01-19 - 2017-07-31 )")
plt.ylabel("Tache Solaires")
plt.grid(True)
```

```python
tache_solaire_mois['13-Month Smoothed Total Sunspot Number'].plot(figsize=(12,8), 
                                                              title= 'Activit des taches solaire moyenne mensuelle', 
                                                              fontsize=12)
plt.xlabel("Series Temporelles (1749-05-31 - 2017-07-31)")
plt.ylabel("Tache Solaires")
plt.grid(True)
```

```python
# creation des donnees X et Y
y = tache_solaire_mois['13-Month Smoothed Total Sunspot Number']
y.dropna(inplace = True)
y.head()
X = [x for x in range(0, len(y))]
print(X)
```

```python
plt.plot(X,y)
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)
scalaire = MinMaxScaler()
mms_y = scalaire.fit_transform(y.values.reshape(-1,1))
plt.plot(X, mms_y)
plt.grid(True)
plt.show()
```

```python
# Separation des ensembles des donnees Test et entrainements
train_size = int(len(mms_y) * 0.67)
test_size = len(mms_y) - train_size
train, test = mms_y[0:train_size,:], mms_y[train_size:len(mms_y)]
print(len(train), len(test))

# creation d'une dataset
def step_ahead(y, jump=1):
    y_current = y[:-(jump + 1), 0]
    y_next = y[jump:-1, 0]
    return np.reshape(y_current, (len(y_current), 1)), np.reshape(y_next, (len(y_next), 1))

# remodeler X = t et Y = t+1

look_back = 1
trainXb, trainYb = step_ahead(train, look_back)
testXb, testYb = step_ahead(test, look_back)

trainXb.shape
type(trainXb[1])

# remodeler les entrees
trainX = np.reshape(trainXb, (trainXb.shape[0], 1, trainXb.shape[1]))
testX = np.reshape(testXb, (testXb.shape[0], 1, testXb.shape[1]))
```

```python
#  Creation et ajustement du model lstm du reseaux de neurones
model = Sequential()

model.add(LSTM(4, input_shape=(1, look_back)))
model.add(Dense(4, activation="sigmoid"))
model.add(Dense(1))
model.summary()
```

```python
and_i_stop = EarlyStopping(monitor = 'loss', patience = 4, mode = 'auto')
model.compile(loss='mean_squared_error', optimizer='adam', metrics=["mae"])
model.fit(trainX, trainYb, epochs=100, callbacks = [and_i_stop], batch_size=1, verbose=2)
```

```python
# Predictions
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)

#inverserpredictions
trainPredict = scalaire.inverse_transform(trainPredict)
trainY = scalaire.inverse_transform(trainYb)
testPredict = scalaire.inverse_transform(testPredict)
testY = scalaire.inverse_transform(testYb)

#calcul l'erreur quadratique moyenne
trainScore = math.sqrt(mean_squared_error(trainY, trainPredict))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY, testPredict))
print('Test Score: %.2f RMSE' % (testScore))
```
